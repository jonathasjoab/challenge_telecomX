# 1. Importação da biblioteca Pandas e visualização inicial dos dados brutos

```python
import pandas as pd
import numpy as np
clientes = pd.read_json('./base_dados/TelecomX_Data.json')
clientes
```

# 2. Normalização do DataSet

```python
customer = pd.json_normalize(clientes['customer'])
customer
```

```python
phone = pd.json_normalize(clientes['phone'])
phone
```

```python
internet = pd.json_normalize(clientes['internet'])
internet
```

```python
account = pd.json_normalize(clientes['account'])
account
```

```python
# Concatenação das colunas originais relevantes com os DataFrames de dados aninhados normalizados
clientes_dados = pd.concat([clientes[['customerID', 'Churn']], customer, phone, internet, account], axis=1)
clientes_dados
```

```python
# Checagem do tipo de dados do dataframe, após normalização
clientes_dados.info()
```

# 3. Verificação de valores inconsistentes

```python
# Verficação de strings vazias e espaços no DataFrame
clientes_dados.isin(['', ' ']).sum().sort_values(ascending=False)
```

```python
# Checando os valores inconsistentes nas colunas 'Charges.Total' e 'Churn'
clientes_dados['Churn'].unique()
```

```python
clientes_dados['Charges.Total'].unique()
```

```python
# Verificação de valores duplicados
clientes_dados.duplicated().sum()
```

```python
# Checando valores nulos
clientes_dados.isna().sum()
```

# 4. Tratamento dos valores inconsistentes

```python
# Padronização das letras das colunas em maiúsculas para minúsculas
clientes_dados.columns = clientes_dados.columns.str.lower()
```
