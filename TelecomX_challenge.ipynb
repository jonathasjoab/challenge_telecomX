# 1. Importação da biblioteca Pandas e visualização inicial dos dados brutos

```python
import pandas as pd
import numpy as np
clientes = pd.read_json('./base_dados/TelecomX_Data.json')
clientes
```

# 2. Normalização do DataSet

```python
customer = pd.json_normalize(clientes['customer'])
customer
```

```python
phone = pd.json_normalize(clientes['phone'])
phone
```

```python
internet = pd.json_normalize(clientes['internet'])
internet
```

```python
account = pd.json_normalize(clientes['account'])
account
```

```python
# Concatenação das colunas originais relevantes com os DataFrames de dados aninhados normalizados
clientes_dados = pd.concat([clientes[['customerID', 'Churn']], customer, phone, internet, account], axis=1)
clientes_dados
```

```python
# Checagem do tipo de dados do dataframe, após normalização
clientes_dados.info()
```

```python
# Quantidade de valores únicos
clientes_dados.nunique()
```

# 3. Verificação de valores inconsistentes

```python
# Verficação de strings vazias e espaços no DataFrame
clientes_dados.isin(['', ' ']).sum().sort_values(ascending=False)
```

```python
# Localiza os valores inválidos ('' ou ' ') na coluna 'Churn' para inspeção antes do tratamento
clientes_dados[clientes_dados['Churn'].isin(['', ' '])][['Churn']]
```

```python
# Localiza os valores inválidos ('' ou ' ') na coluna 'Charges.Total' para inspeção antes do tratamento
clientes_dados[clientes_dados['Charges.Total'].isin(['', ' '])][['Charges.Total']]
```

```python
# Verificação de valores duplicados
clientes_dados.duplicated().sum()
```

```python
# Checando valores nulos
clientes_dados.isnull().sum()
```

# 4. Tratamento dos valores inconsistentes

```python
# Aplicação de valores nulos na coluna 'Churn'
clientes_dados['Churn'] = clientes_dados['Churn'].replace('', np.nan)

# Visualização dos valores nulos
clientes_dados['Churn'].isnull().sum()
```

```python
# Aplicação de valores nulos na coluna 'Charges.Total'
clientes_dados['Charges.Total'] = clientes_dados['Charges.Total'].replace(' ', np.nan)

# Visualização dos valores nulos
clientes_dados['Charges.Total'].isnull().sum()
```

```python
# Deletando as linhas com valores nulos
clientes_dados = clientes_dados.dropna(subset=['Churn'])
clientes_dados = clientes_dados.dropna(subset=['Charges.Total'])
```

```python
# Nova verificação de valores vazios e nulos no DataFrame
clientes_dados.isin(['', ' ']).sum().sort_values(ascending=False)
```

```python
clientes_dados.isnull().sum()
```

```python
#  Conversão do coluna Charges.Total do tipo 'object' para 'float'
clientes_dados['Charges.Total'] = clientes_dados['Charges.Total'].astype(np.float64)
```
