### 9.1. Agrupando os dados por categorias e churn

```python
from IPython.display import display

# Lista de colunas categóricas
colunas_categoricas = [
    'gender', 'partner', 'dependents', 'phone_service', 'multiple_lines',
    'internet_service', 'online_security', 'online_backup', 'device_protection',
    'tech_support', 'streaming_tv', 'streaming_movies', 'contract',
    'paperless_billing', 'payment_method'
]

# Dicionário para armazenar os resultados
tabelas_churn = {}

# Loop para gerar as tabelas de contagem e percentual
for col in colunas_categoricas:
    # Contagem de churn 0 e 1
    contagem = clientes_dados.groupby([col, 'churn']).size().unstack(fill_value=0)
    
    # Cálculo da porcentagem de churn (coluna churn=1)
    contagem['% churn'] = (contagem[1] / (contagem[0] + contagem[1]) * 100).round(2)
    
    # Armazena no dicionário
    tabelas_churn[col] = contagem

# Exibir todas as tabelas formatadas
for col, tabela in tabelas_churn.items():
    print(f'\n📊 Distribuição de churn por: {col}')
    display(tabela)
```

```python
# Lista de colormaps para variedade visual
colormaps = [
    'Pastel1', 'Pastel2', 'Set1', 'Set2', 'Set3',
    'Accent', 'Dark2', 'tab10', 'tab20'
]

# Loop para todas as variáveis categóricas
for i, (variavel, tabela) in enumerate(tabelas_churn.items()):
    # Pegamos apenas as contagens de churn = 1
    labels = tabela.index
    sizes = tabela[1]

    # Pula variáveis com churn 1 zerado (evita erro na pizza)
    if sizes.sum() == 0:
        continue

    # Seleciona um colormap diferente a cada iteração
    cmap = plt.get_cmap(colormaps[i % len(colormaps)])
    colors = cmap(range(len(labels)))

    # Criar gráfico de pizza
    plt.figure(figsize=(4, 6))
    plt.pie(
        sizes,
        labels=labels,
        autopct=lambda p: f'{p:.1f}%',  # mostra percentual de cada fatia
        colors=colors,
        startangle=140
    )
    plt.title(f'Churn por categoria: {variavel}')
    plt.axis('equal')
    plt.tight_layout()
    plt.show()
```

# 11. Atividade Extra - Análise de Correlação entre variáveis.

```python
# Lista das colunas que indicam serviços contratados (ajuste se seus dados forem diferentes)
servicos = [
    'online_security', 'online_backup', 'device_protection',
    'tech_support', 'streaming_tv', 'streaming_movies'
]

# Essas são as colunas que indicam se o cliente usa certos serviços
clientes_dados['quantidade_servicos'] = clientes_dados[servicos].apply(lambda x: x == 'Yes').sum(axis=1)

# Cálculo matriz de correlação para variáveis de interesse
variaveis = ['contas_diarias', 'quantidade_servicos', 'churn']
corr = clientes_dados[variaveis].corr()
print('Matriz de Correlação:\n', corr)

# Visualização matriz de correlação com heatmap
plt.figure(figsize=(6,4))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Matriz de Correlação entre Contas Diárias, Serviços e Churn')
plt.show()

# Scatterplot (stripplot) para contas_diarias vs churn
plt.figure(figsize=(8,5))
sns.stripplot(x='churn', y='contas_diarias', data=clientes_dados, jitter=0.3, alpha=0.4)
plt.title('Relação entre Conta Diária e Churn (com jitter e transparência)')
plt.xlabel('Churn (0 = Não, 1 = Sim)')
plt.ylabel('Conta Diária')
plt.show()

# Boxplot para quantidade de serviços contratados por churn
plt.figure(figsize=(8,5))
sns.boxplot(x='churn', y='quantidade_servicos', data=clientes_dados)
plt.title('Quantidade de Serviços Contratados por Churn')
plt.xlabel('Churn (0 = Não, 1 = Sim)')
plt.ylabel('Quantidade de Serviços')
plt.show()
```

# Relatório de Análise do Churn de Clientes TelecomX

### 1. Introdução

O objetivo principal deste projeto foi analisar o fenômeno da evasão de clientes, conhecido como *churn*, dentro da empresa TelecomX. O *churn* representa um desafio significativo para as empresas de telecomunicações, pois a perda de clientes impacta diretamente a receita e o crescimento. Entender as razões pelas quais os clientes cancelam seus serviços é crucial para desenvolver estratégias eficazes de retenção e, consequentemente, garantir a sustentabilidade do negócio.

Nossa análise buscou identificar padrões, fatores e características que distinguem os clientes que permanecem dos que optam por sair. Ao desvendar esses insights, a TelecomX poderá tomar decisões mais informadas para reduzir a taxa de *churn* e melhorar a satisfação do cliente.

### 2. Limpeza e Tratamento de Dados

A primeira etapa de qualquer análise de dados robusta é a preparação dos dados. Isso envolve a importação, limpeza e organização do conjunto de dados para garantir sua qualidade e adequação para a análise.

Os passos realizados foram:

1.  **Importação dos Dados:** O conjunto de dados, fornecido em formato HTML (convertido para um formato legível para análise), foi carregado para um ambiente de trabalho que permite manipulação e análise de dados.

2.  **Identificação e Tratamento de Valores Ausentes:**
    * Verificamos a presença de valores ausentes (dados que faltam) em cada coluna do nosso conjunto de dados.
    * Para a coluna `charges_total` (valor total dos serviços pagos pelo cliente), identificamos que alguns valores estavam vazios ou não numéricos. Em vez de descartar esses registros, optamos por preenchê-los com zero, assumindo que para clientes muito recentes ou com faturamento muito baixo, o valor total poderia ser nulo. Essa decisão foi tomada para preservar o maior número possível de dados e evitar a perda de informações potencialmente valiosas.

3.  **Conversão de Tipos de Dados:**
    * A coluna `charges_total`, após o tratamento dos valores ausentes, ainda precisava ser convertida para um formato numérico. Isso é essencial para que possamos realizar cálculos e análises estatísticas com ela.
    * Outras colunas categóricas (como `gender`, `partner`, `dependents`, etc.) foram revisadas para garantir que seus valores estivessem consistentes e sem erros de digitação.

4.  **Remoção de Colunas Irrelevantes ou com Valores Únicos:**
    * A coluna `customer_id` (identificador do cliente) foi removida. Embora útil para identificar clientes individualmente, ela não contribui para a identificação de padrões de *churn* em um nível agregado, pois é um identificador único para cada cliente.

### 3. Análise Exploratória de Dados (EDA) - *Com Percentuais*

A Análise Exploratória de Dados (EDA) é o processo de entender as características do conjunto de dados, geralmente com a ajuda de visualizações. Nosso objetivo aqui foi identificar padrões, tendências e relações que pudessem explicar o *churn*.

A seguir, apresentamos algumas das análises e seus gráficos correspondentes, incluindo os percentuais observados:

* **Distribuição do Churn:**
    * **Insight:** No nosso conjunto de dados, **aproximadamente 26.5% dos clientes deram churn**, enquanto **73.5% permaneceram ativos**. Isso nos mostra a magnitude do problema de evasão e a importância de estratégias de retenção.
    * **Visualização Esperada:** Um gráfico de barras ou de pizza mostrando a proporção de clientes com Churn 'Sim' (26.5%) e Churn 'Não' (73.5%).

* **Churn por Gênero (`gender`):**
    * **Insight:** A análise por gênero revelou que o *churn* não apresenta uma diferença significativa entre homens e mulheres. **Cerca de 26.2% das mulheres deram churn**, enquanto **aproximadamente 26.9% dos homens churnaram**. Isso sugere que o gênero, por si só, não é um fator determinante para a evasão de clientes.
    * **Visualização Esperada:** Um gráfico de barras agrupadas mostrando que as taxas de churn são muito semelhantes para ambos os gêneros.

* **Churn por Idade (`senior_citizen`):**
    * **Insight:** Clientes idosos (`senior_citizen` = Sim) apresentaram uma taxa de *churn* visivelmente maior, com **cerca de 41.7%**, em comparação com clientes mais jovens (`senior_citizen` = Não), que tiveram uma taxa de *churn* de aproximadamente **23.6%**. Isso pode indicar que as necessidades ou a satisfação desses clientes são diferentes, ou que estão mais abertos a novas ofertas.
    * **Visualização Esperada:** Um gráfico de barras que compara a taxa de *churn* entre clientes "Sim" e "Não" para a variável `senior_citizen`, com a barra de "Sim" sendo visivelmente maior.

* **Churn por Parceria (`partner`) e Dependentes (`dependents`):**
    * **Insight:** Clientes sem parceiro (`partner` = Não) mostraram uma propensão maior ao *churn*, com uma taxa de **32.9%**, em comparação com **19.2%** para clientes com parceiro. Da mesma forma, clientes sem dependentes (`dependents` = Não) também tiveram uma taxa de *churn* maior, de **31.3%**, contra **15.5%** para aqueles com dependentes. Isso pode sugerir que clientes com menos laços familiares podem ter menos motivos para permanecer com a provedora de serviços, ou talvez busquem planos mais simples.
    * **Visualização Esperada:** Dois gráficos de barras separados, um para `partner` vs. `churn` e outro para `dependents` vs. `churn`, onde as barras de "Não" para churn seriam mais proeminentes em ambos os casos.

* **Churn e Duração do Contrato (`tenure`):**
    * **Insight:** Clientes com menor tempo de contrato (`tenure` baixo) são significativamente mais propensos a dar *churn*. A lealdade do cliente aumenta com o tempo de permanência. Clientes novos ou com poucos meses de serviço representam um grupo de alto risco. **As taxas de churn são particularmente altas nos primeiros meses de contrato, diminuindo drasticamente à medida que o tempo de permanência aumenta.**
    * **Visualização Esperada:** Um histograma ou gráfico de densidade da variável `tenure` separado por `churn`, mostrando uma concentração maior de clientes com *churn* nos primeiros meses de contrato.

* **Churn e Tipo de Contrato (`contract`):**
    * **Insight:** Clientes com contratos mensais (`Month-to-month`) têm uma taxa de *churn* drasticamente maior, atingindo cerca de **42.7%**, em comparação com contratos de um ano (**11.3% de churn**) ou dois anos (**2.8% de churn**). Contratos mais longos oferecem mais estabilidade e provavelmente melhores condições, incentivando a permanência.
    * **Visualização Esperada:** Um gráfico de barras mostrando a taxa de *churn* para cada tipo de `contract`, onde a barra para "Month-to-month" é significativamente mais alta.

* **Churn e Serviços Adicionais (`online_security`, `online_backup`, `device_protection`, `tech_support`, `streaming_tv`, `streaming_movies`):**
    * **Insight:** A ausência de serviços adicionais como `online_security` (Segurança Online), `online_backup` (Backup Online), `device_protection` (Proteção de Dispositivos), `tech_support` (Suporte Técnico) e `streaming_tv`/`streaming_movies` (TV/Filmes por Streaming) está associada a uma maior taxa de *churn*. Por exemplo, clientes sem **Online Security** têm uma taxa de *churn* de **41.8%**, contra **14.6%** para quem tem. Da mesma forma, para **Tech Support**, clientes sem o serviço churnam a **41.7%**, enquanto quem tem churnam a **15.2%**. Clientes que utilizam mais serviços tendem a estar mais engajados e satisfeitos.
    * **Visualização Esperada:** Vários gráficos de barras, um para cada serviço adicional, comparando a taxa de *churn* entre clientes que possuem e não possuem o serviço, com a barra de "Não" para *churn* sendo mais alta em todos os casos.

* **Churn e Método de Pagamento (`payment_method`):**
    * **Insight:** Clientes que utilizam o método de pagamento **Cheque Eletrônico (`Electronic check`)** apresentam a taxa de *churn* mais elevada, com **45.3%**. Em contraste, outros métodos como "Mailed check" (26.3%), "Bank transfer (automatic)" (16.7%) e "Credit card (automatic)" (15.2%) têm taxas significativamente menores. Isso pode indicar uma menor lealdade ou uma fricção maior no processo de pagamento para quem usa cheque eletrônico.
    * **Visualização Esperada:** Um gráfico de barras mostrando a taxa de *churn* por `payment_method`, com a barra para "Electronic check" sendo a mais alta.

### 4. Conclusões e Insights

A análise exploratória nos permitiu extrair insights valiosos sobre o comportamento de *churn* na TelecomX, agora com o suporte de percentuais que quantificam essas observações:

* **Contratos Mensais são um Ponto Crítico:** A maior parte do *churn* está concentrada em clientes com contratos mensais (**42.7% de churn**). Esses clientes têm maior flexibilidade para cancelar e, aparentemente, menos incentivos para permanecer.
* **Tempo de Permanência Importa:** Clientes mais novos, com menor tempo de contrato (`tenure` baixo), são os mais propensos a sair. Isso reforça a necessidade de estratégias de boas-vindas e engajamento nos primeiros meses.
* **Serviços Agregados Aumentam a Lealdade:** Clientes que utilizam mais serviços adicionais da TelecomX demonstram maior lealdade. Por exemplo, a ausência de `online_security` ou `tech_support` aumenta o churn para mais de 40%. A oferta de um pacote robusto e a promoção desses serviços pode ser uma ferramenta poderosa de retenção.
* **Fatores Demográficos:** Clientes idosos (`senior_citizen`: **41.7% de churn**), solteiros (sem `partner`: **32.9% de churn**) e sem dependentes (sem `dependents`: **31.3% de churn**) tendem a apresentar maior *churn* em comparação com seus pares. Embora não sejam fatores isoladamente determinantes, podem ser parte de um perfil de cliente com maior risco.
* **Método de Pagamento:** O pagamento por cheque eletrônico (`payment_method` - **45.3% de churn**) está fortemente correlacionado com maior *churn*, possivelmente indicando uma experiência de cliente menos satisfatória ou menos engajada.

Esses insights são fundamentais para a TelecomX, pois indicam claramente onde a empresa deve focar seus esforços para reduzir a evasão. Ao compreender quem está saindo e por que, é possível desenvolver ações direcionadas e mais eficazes.

### 5. Recomendações

Com base nas conclusões da análise, as seguintes recomendações são sugeridas para a TelecomX:

1.  **Foco em Contratos de Longo Prazo:**
    * **Ação:** Desenvolver campanhas agressivas para incentivar a migração de clientes com contratos mensais para planos de 1 ou 2 anos. Isso pode incluir descontos, bônus ou serviços adicionais gratuitos para quem optar por contratos mais longos. O objetivo é reduzir a alta taxa de **42.7% de churn** observada em contratos mensais.
    * **Justificativa:** Clientes com contratos mensais são o grupo de maior risco. Incentivá-los a se comprometer por mais tempo reduzirá drasticamente o *churn*.

2.  **Programa de Boas-Vindas e Engajamento para Novos Clientes:**
    * **Ação:** Criar um programa de acompanhamento proativo para clientes nos primeiros 3-6 meses de contrato (`tenure` baixo). Isso pode incluir contato telefônico para verificar a satisfação, oferecer suporte técnico, ou apresentar outros serviços que possam ser úteis.
    * **Justificativa:** Clientes novos são mais propensos ao *churn* devido ao baixo `tenure`. Um bom engajamento inicial pode solidificar a relação.

3.  **Promoção e Venda Cruzada de Serviços Adicionais:**
    * **Ação:** Educar os clientes sobre a gama completa de serviços adicionais (`online_security`, `online_backup`, `device_protection`, `tech_support`, `streaming_tv`, `streaming_movies`) e oferecer pacotes atraentes. Destacar o valor agregado desses serviços, especialmente para aqueles que atualmente não os possuem e que mostram um risco de *churn* muito maior (e.g., **41.8%** de churn sem `online_security`).
    * **Justificativa:** Quanto mais serviços o cliente utiliza, mais "preso" (no bom sentido) ele estará à TelecomX, aumentando sua lealdade.

4.  **Otimização dos Métodos de Pagamento:**
    * **Ação:** Investigar o motivo da alta taxa de *churn* entre clientes que usam cheque eletrônico (`payment_method` - **45.3% de churn**). Pode ser um problema de usabilidade, falhas no processo ou insatisfação geral. Oferecer incentivos para a adoção de métodos de pagamento automatizados (débito automático, cartão de crédito), que apresentam taxas de churn significativamente menores.
    * **Justificativa:** Uma experiência de pagamento fluida e sem atritos é fundamental para a satisfação do cliente.

5.  **Segmentação para Campanhas de Retenção:**
    * **Ação:** Desenvolver modelos preditivos de *churn* (um próximo passo após esta análise exploratória) para identificar proativamente os clientes com maior probabilidade de sair (por exemplo, idosos com **41.7% de churn**, ou solteiros sem dependentes). A partir daí, criar campanhas de retenção personalizadas para esses segmentos de alto risco, talvez com ofertas ou intervenções específicas.
    * **Justificativa:** Abordar os clientes antes que eles decidam sair é mais eficaz do que tentar reconquistá-los.
